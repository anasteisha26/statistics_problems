```python
import pandas as pd
import numpy as np
import random
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
city_a = [1] * 200 + [0] * 9800
city_b = [1] * 230 + [0] * 9770
altogether = city_a + city_b

n = 10000
```

```python
print(sum(city_a)/len(city_a))
print(sum(city_b)/len(city_b))
print(sum(city_b)/len(city_b) - sum(city_a)/len(city_a))
```

```python
def generate():
    random_selection_a = random.sample(altogether, n)
    a = random_selection_a.count(1)
    b = 430 - random_selection_a.count(1)
    diff = (max(a, b) - min(a,b))/10000
    return diff
```

```python
list_all = []

for i in range (5000):
    list_all.append(generate())
```

```python
values = np.array(list_all)

plt.figure(figsize=(20,5))
sns.histplot(values, bins=len(set(values)), kde=True, color="blue")


plt.xlabel("Number")
plt.ylabel("Count")
plt.title("Number Distribution")

plt.show()
```

```python
list_2 = []

for i in range (5000):
    list_2.append(generate_negs())
```
